```python
import numpy as np
import pandas as pd

# 從pd.read_csv裡看到讀取的檔案是利用\t分割，利用sep=''更改格式
df=pd.read_csv('./marketing_campaign.csv',sep='\t')

# 計算資料中有多少空值
df.isnull().sum()

# 刪除Missing Values
df=df.dropna()

# 刪除ID
df.drop(columns=['ID'], inplace = True)

df
```
