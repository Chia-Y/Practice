```python
# 刪除ID、Dt_Customer、Response(原先別人已作的結果)
# inplace=False 創建新的資料對它進行修改 =True 直接對原資料修改
df2 = df.drop(columns=['ID', 'Dt_Customer', 'Education', 'Marital_Status', 'Response'], inplace = False)

# axis=0指row，axis=1指columns
# df2 = df.drop(['ID', 'Dt_Customer'], axis = 1, inplace = False) 另一種寫法

df2
```

```python
from sklearn.model_selection import train_test_split

X = df2
y = df.Response

# random_state隨機數種子：該組隨機數的編號，在需要重複試驗的時候，保證得到一組一樣的隨機數
# 比如每次都填1，其他參數一樣的情況下你得到的隨機數組是一樣的。但填0或不填，每次都會不一樣。

# 要加上random_state(數字隨便設)，每次抓的資料才會是一樣的
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 4)
print(X_train)
print('====')
print(X_test)
print('====')
print(y_train)
print('====')
print(y_test)
print('====')
```

```python
from sklearn.tree import DecisionTreeClassifier

# 要加上random_state，每次抓的資料才會是一樣的
model = DecisionTreeClassifier(random_state = 4)
model.fit(X_train, y_train)

# 預測百分比(比較符合實際操作)
y_pred = model.predict_proba(X_test)[:,1] # 先選row再選columns，冒號:代表row全選，1代表只要第一個culomns

# 預測T/F
# y_pred = model.predict(X_test)

y_pred
```

```python
# auc
from sklearn.metrics import roc_auc_score
auc = roc_auc_score(y_test, y_pred)
auc
```

```python
# 計算準確率
sum(y_pred == y_test) / len(y_test)
```
